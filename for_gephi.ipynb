```python
import numpy as np
import pandas as pd
import string
from collections import Counter
import networkx as nx
import matplotlib.pyplot as plt 


def file_open(path):
    print('Start read data from file')
    input_read = pd.read_excel(path,skip_blank_lines=True) 
    df = pd.DataFrame(input_read)
    print('Read data')
    return df
```

```python
def relation_data(path):
    res = []
    df = file_open(path)
      
    for index,data in enumerate(df["issue_ID"],start=1):
        if index == df.shape[0]:
            break
        if df["issue_ID"][index]== df["issue_ID"][index-1]:
            res.append((df["commenter"][index-1],df["commenter"][index]))
    return res
```

```python

path = "data/closed_comment.xlsx"
data = relation_data(path)
c_directed = Counter(data)    #directed relation  ex: [a,b] != [b,a]
[sorted(tuple(i)) for i in data]
```

```python
c_Undirected = Counter([tuple(sorted(i)) for i in data])   #undirected relation  ex: [a,b] == [b,a]
```

```python
w = [(key[0],key[1],value) for key,value in c_Undirected.items()]
```

```python
a = []
b = []
c = []
for i in w:
    a.append(i[0])
    b.append(i[1])
    c.append(i[2])
    
dit = {'Source':a, 'Target':b,'Weight':c}
df = pd.DataFrame(dit)
df.to_csv(r'output_for_gephi2',columns=['Source','Target','Weight'],index=False,sep=',')
```
